### Importing Libraries and Inspecting the Data

So it turns out that the labels for this data are a bit more broad than the title states. That is fine in any case. Good to know from the start.  

Now using Keras, lets create a training set and a validation set.

```python
train_dir = 'seg_train'

train_datagen = ImageDataGenerator(rescale = 1./255, validation_split=.25)

train_generator = train_datagen.flow_from_directory(
                train_dir,
                target_size = (150, 150),
                batch_size = 20,
                subset = "training",
                seed = 2902,
                class_mode = 'categorical')

valid_generator = train_datagen.flow_from_directory(
                train_dir,
                target_size = (150, 150),
                batch_size = 20,
                subset = "validation",
                class_mode = 'categorical')
```

I have a total of 12777 images for my training set and 4257 for my validation set.

## Base Model

Taking a simple deep learning architecture, I will run through different number of layers while altering the layer sizes, and then using different numbers of convolutional layers. I will also use Tensorboard to compare the accuracy and loss from the training data and the validation data.

```python
dense_layers = [0, 1, 2]
layer_sizes = [64, 128, 256]
conv_layers = [1, 2, 3]
```

```python
for dense_layer in dense_layers:
    for layer_size in layer_sizes:
        for conv_layer in conv_layers:
            
            NAME = f"{conv_layer}-conv-{layer_size}-nodes-{dense_layer}-dense-{int(time.time())}"
            print(NAME)
            
            model = Sequential()
            
            model.add(Conv2D(32, (3, 3), activation = 'relu', input_shape = (150, 150, 3)))
            model.add(MaxPooling2D((2,2)))
            
            for l in range(conv_layer - 1):
                model.add(Conv2D(64, (3, 3), activation='relu'))
                model.add(MaxPooling2D((2, 2)))

            model.add(Flatten())
            
            for l in range(dense_layer):
                model.add(Dense(layer_size, activation='relu'))
                model.add(Dense((layer_size + 128), activation='relu'))
            
            model.add(Dense(6, activation='sigmoid'))
            
            tensorboard = TensorBoard(log_dir="logs/{}".format(NAME))  #Logs the training and validation loss and accuracy


            model.compile(loss='categorical_crossentropy',
                          optimizer='adam',
                          metrics=['acc'])

            model.fit_generator(
                  train_generator,
                  steps_per_epoch=100,
                  validation_data = valid_generator, 
                  epochs=10,
                  callbacks = [tensorboard])
```

## Model V1.5 - Reduced Batch/Epoch Size

Creating a new train generator with a decreased 

```python
train_generatorV2 = train_datagen.flow_from_directory(
                train_dir,
                target_size = (150, 150),
                batch_size = 13,
                seed = 2902,
                subset = "training",
                class_mode = 'categorical')

valid_generatorV2 = train_datagen.flow_from_directory(
                train_dir,
                target_size = (150, 150),
                batch_size = 13,
                seed = 2902, 
                subset = "validation",
                class_mode = 'categorical')
```

I retained the same amount of images but reduced the batch size of each set.

I am going to rerun my previous architecture with a lower steps per epoch in hopes to figure out why I am getting NAN values. Also I am sticking with 3 convolutional layers owning to the fact they gave me the best results.

```python
dense_layers = [0, 1, 2]
layer_sizes = [64, 128, 256]
conv_layers = [3]
```

```python
for dense_layer in dense_layers:
    for layer_size in layer_sizes:
        for conv_layer in conv_layers:
            
            NAME = f"13Batchsize-{conv_layer}-conv-{layer_size}-nodes-{dense_layer}-dense-{int(time.time())}"
            print(NAME)
            
            model = Sequential()
            
            model.add(Conv2D(32, (3, 3), activation = 'relu', input_shape = (150, 150, 3)))
            model.add(MaxPooling2D((2,2)))
            
            for l in range(conv_layer - 1):
                model.add(Conv2D(64, (3, 3), activation='relu'))
                model.add(MaxPooling2D((2, 2)))

            model.add(Flatten())
            
            for l in range(dense_layer):
                model.add(Dense(layer_size, activation='relu'))
                model.add(Dense((layer_size + 128), activation='relu'))
            
            model.add(Dense(6, activation='sigmoid'))
            
            tensorboard = TensorBoard(log_dir="logs/{}".format(NAME))


            model.compile(loss='categorical_crossentropy',
                          optimizer='adam',
                          metrics=['acc'])

            model.fit_generator(
                  train_generatorV2,
                  steps_per_epoch=75,
                  validation_data = valid_generatorV2, 
                  epochs=10,
                  callbacks = [tensorboard])
```

The lack of dense layers seem to be at the heart of the NAN values I was receiving. My validation accuracy is still around 70%. I am going to add in some more dense layers. Then some dropout layers to get my test accuracy more in line with my validation accuracy.

```python
layer_sizes = [64, 128, 256]
conv_layers = [3]
```

```python
for layer_size in layer_sizes:
        for conv_layer in conv_layers:
            
            NAME = f"V2-{conv_layer}-conv-{layer_size}-nodes-{conv_layer}-dense-{int(time.time())}"
            print(NAME)
            
            model = Sequential()
            
            model.add(Conv2D(32, (3, 3), activation = 'relu', input_shape = (150, 150, 3)))
            model.add(MaxPooling2D((2,2)))
            
            for l in range(conv_layer - 1):
                model.add(Conv2D(64, (3, 3), activation='relu'))
                model.add(MaxPooling2D((2, 2)))
                model.add(Dropout(.25, seed=2902))
                model.add(Dense(layer_size, activation='relu'))

            model.add(Flatten())
            
            model.add(Dense(6, activation='sigmoid'))
            
            tensorboard = TensorBoard(log_dir="logs/{}".format(NAME))


            model.compile(loss='categorical_crossentropy',
                          optimizer='adam',
                          metrics=['acc'])

            model.fit_generator(
                  train_generatorV2,
                  steps_per_epoch=75,
                  validation_data = valid_generatorV2, 
                  epochs=10,
                  callbacks = [tensorboard])
```

# Pre-Trained Models

In transfer learning, we take the pre-trained weights of an already trained model (one that has been trained on millions of images belonging to 1000s of classes, on several high power GPUs for several days) and use these already learned features to predict new classes. It requires some changes towards the end of the trained model, along with an array of other hyperparameters, to optimize the model to fit the new data.

## ResNet

Making use of Keras extensive library, lets try `ResNet` which we imported earlier.

ResNet is a short name for a residual network. Microsoft introduced a deep residual learning framework. Instead of hoping every few stacked layers directly fit a desired underlying mapping, they explicitly let these layers fit a residual mapping.

Now lets import the weights of ResNet and get to work.

```python
pre_trained_model = ResNet50(weights='imagenet', include_top = False)
```

When using a pre-trained model, you are required to change the final layer to match your dataset. In my cass, I have 6 classes therefore I need to make a final layer that returns only six possible values. Furthermore, I am going to add a few more dense layers before my output layer.

```python
x = pre_trained_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024,activation='relu')(x) #we add dense layers so that the model can learn more complex functions and classify for better results.
x = Dense(1024,activation='relu')(x) #dense layer 2
x = Dense(512,activation='relu')(x) #dense layer 3
preds = Dense(6,activation='softmax')(x) #final layer with softmax activation
```

```python
model=Model(inputs=pre_trained_model.input,outputs=preds)
```

Lets check out the network itself and see how deep it is. 

```python
for i,layer in enumerate(model.layers):
    print(i,layer.name)
```

ResNet is a much bigger network than the one I created myself. There are a few more steps to go through before implementing this new model though.

As mentioned above there a hyperparameters to play with to improve the effectiveness of using a pre-trained model. The one I will implement for the rest of the notebook is freezing layers.  

Freezing a layer prevents its weights from being modified. This means we can leverage not only the model architecture itself but also what is has learned previously. 

As you can see, ResNet50 has ~180 layers. It looks like some activation functions are labeled as layer within themselves, but to call it one would be in error. Nevertheless, I will retain all the weights of the network and see if I can't get some good accuracy.

### ResNet50 Full Forzen Layers

```python
for layer in model.layers: #Freezing all the weights in the network
    layer.trainable=False
```

```python
NAME = f"ResNet50-imagenet-ALL-frozen-layers-{int(time.time())}"

tensorboard = TensorBoard(log_dir="logs/{}".format(NAME))

model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])

model.fit_generator(
                  train_generatorV2,
                  steps_per_epoch=100,
                  validation_data = valid_generatorV2,
                  validation_steps= len(valid_generatorV2)/13,
                  epochs=15,
                  callbacks = [tensorboard])
```

### Partial Forzen Layers

Making a quick for loop to freeze a certain subset of layers. I arbitrarily picked 20 layers to freeze.

```python
for layer in model.layers[:20]:
    layer.trainable=False
for layer in model.layers[20:]:
    layer.trainable=True    
```

Rerunning the model.

```python
NAME = f"ResNet50-imagenet-20-frozen-layers-{int(time.time())}"

tensorboard = TensorBoard(log_dir="logs/{}".format(NAME))

model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])

model.fit_generator(
                  train_generatorV2,
                  steps_per_epoch=100,
                  validation_data = valid_generatorV2,
                  validation_steps= len(valid_generatorV2)/13,
                  epochs=15,
                  callbacks = [tensorboard])
```

### VGG16 - How many to freeze?

VGG16 is smaller model than ResNet, that is my motivation of chosing one over the other. So first start by bringing in a new model that is unedited.

```python
pre_trained_VGG16 = VGG16(weights='imagenet', include_top = False)
```

Sticking with what I did with the other trials, append the a last few layers and my final classifier.

```python
x = pre_trained_VGG16.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024,activation='relu')(x) #we add dense layers so that the model can learn more complex functions and classify for better results.
x = Dense(1024,activation='relu')(x) #dense layer 2
x = Dense(512,activation='relu')(x) #dense layer 3
preds = Dense(6,activation='softmax')(x) #final layer with softmax activation
```

```python
model=Model(inputs=pre_trained_VGG16.input,outputs=preds)
```

With our model compiled, lets print out the layer and its name.

```python
for i,layer in enumerate(model.layers):
    print(i,layer.name)
```

Knowing the name of all the layers and the depth of the archiecture, I will run 18 new networks freezing a new layer every time it runs through.

```python
for i in range(0,19):
    
    #Standard model 
    pre_trained_VGG16 = VGG16(weights='imagenet', include_top = False)
    x = pre_trained_VGG16.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024,activation='relu')(x)
    x = Dense(1024,activation='relu')(x) 
    x = Dense(512,activation='relu')(x) 
    preds = Dense(6,activation='softmax')(x)
    model=Model(inputs=pre_trained_VGG16.input,outputs=preds)
                                          
    #Loop for freezing layers                                      
    for layer in model.layers[:i]:
        layer.trainable=False
    for layer in model.layers[i:]:
        layer.trainable=True    
        
        
    NAME = f"VGG16-{i}-layers-frozen-{int(time.time())}"

    tensorboard = TensorBoard(log_dir="logs/{}".format(NAME))

    model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])

    model.fit_generator(
                      train_generatorV2,
                      steps_per_epoch=100,
                      validation_data = valid_generatorV2,
                      validation_steps= len(valid_generatorV2)/13,
                      epochs=15,
                      callbacks = [tensorboard])
```
